# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
#%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(6,6))    # We've created a figure and We've created an axes

# 3. Plot the data
ax.plot(x, y)    # and now what we're doing is adding data to our axes

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")

plt.show()
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()    # we are going to create a figure and an axis. we write plt.subplots because this is the method that we're going to be using
ax.plot(x, x**2);   # let's do x to the power of 2 and let's see what happens
plt.show()
```

```python
# Use the same data to make a Scatter Plot
fig, ax = plt.subplots()    # if we want to create another plot we have to always reset fig and ax if we run a different cell
ax.scatter(x, np.exp(x)) # if we want to create a scatter plot we need to call scatter method on ax (shift tab)
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
# this is a sine wave. Seno trigonométrico, por elementos. para cada elemento del arreglo x devuelve sin x, por lo tanto creamos un vector y
# con las componentes de seno de x de manera discreta, al plotear con el metodo scatter plot estamos obteniendo un grafico de dispersión de 
# la función trigonometrica
# shift tab
```

### Two options for subplots

```python
# Subplot option 1 (this is going to look a bit different from what we've been doing)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
plt.show()
```

```python
# subplots crea una figura y un conjunto de subplots.
# nrows, ncols: number of rows/columns of the subplot grid (shift tab)

# So we've got a fig and then a tuple which is of tuples ax1, ax2 and ax3, ax4
# and then in subplots we have key arguments nrows =2, ncols=", figsize=(10,5)

# And we obtained two rows of subplots, two columns of subplots
# ax1 corresponds to the upper left subplot
# ax2 corresponds to the bottom left subplot
# ax3 corresponds to the upper right subplot
# ax4 corresponds to the bottom right subplot
```

```python
# Subplots option 2 (less common) we stick to option 1
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));

plt.show()
```

### Which one should you use? (Pyplot vs Matplotlib Object Oriented method?)

* When plotting something quickly, it's okay to use the Pyplot method
* When plotting something more advanced, use the Object Oriented Method

```python
# now we're going to plot the histograms using the Object Oriented style
# Let's say we want to explore the patients which are over the age of 50
# We only want to look at those patients and we want to create a little bit of a plot and start to do some data analysis on those
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
# we can acces those by using Boolean indexing
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target');
```

```python
# We have chol on the y-axis (the vertical axis), and age on the x axis (the horizontal axis)
# and c is giving us this colorbar to show us if someone has heart disease (1 black) or they don't (0 white)

# OO method mixed with pyplot method
# we're going to redo the same plot as before
fig, ax = plt.subplots(figsize=(10, 6))    # adjust the size of the figure with fig size (a width of 10 inches and a height o 6 inches) 1"=25.4mm
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
# we want you to plot this data (x, y and c on the ax object we ha0ve just created)
# Axe is this part here (orange/pink rectangle in the Anatomy of a Matplotlib plot) where we're adding data
# whereas Figure is the light blue rectangle that contains the Axes.
#ax.set_xlim([45, 100]);    # delimita los extremos de la escala graduada en el eje x 
# (queremos ver edades de 45 a 100 en el eje horizontal ni mas ni menos) 

# This is going to change the limitations of out X axis to be from 45 to 100 
```

```python
over_50.target.values
# 1 tiene enfermedad cardiaca
# 0 no tiene enfermedad cardiaca
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol level",
       xlabel="Age",
       ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(), title="Target"); 
# this little star * is saying, hey, grab the values from scatter and unpack the legend elements
# which is going to query this c parameter here and loock back up at the heart_disease dataframe and go,
# what values are in this, in the target column, so we've got ones and zeros and then it's going to create
# the legend built off those elements.

# it just goes into the scatter variable
# and the it's going to find the legend elements (which it gets from this C parameter)
# which it then looks back up into out heart disease data frame
# it is going to look at the target column (ones and zeros) 
# and then it's going to color the legend the same as the values here in the plot

# What if we wanted something like a horizontal line across here to indicate the average cholesterol?

# Añade una línea horizontal a través del eje.

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
# we are going to combine the previous plot with the feature thalach (which is max heart rate)
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True);
# now we are creating a single figure, but it's going to have two axes on it (areas where points can be specified in terms of x-y coordinates)
# Axeses are the regions of the figure where you can add data that usually contains two axis objets

# Add some data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
# if you add sharex=True we can remove the xlabel="Age" argument

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"], 
                       y=over_50["thalach"],
                       c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Maximum Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# So what this has done is it's telling our plot to go that our subplots here have the same X values
# Right now i'm starting to see just how powerful Matplotlib in the Object Oriented API can be
# try a histogram in the 1st subplot and the a bar graph in the 2nd subplot
```

```python
# Constrain the limits of the axis
# Customizing the y and x axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True);
# now we are creating a single figure, but it's going to have two axes on it (areas where points can be specified in terms of x-y coordinates)
# Axeses are the regions of the figure where you can add data that usually contains two axis objets

# Add some data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

ax0.set_xlim([50, 80])

# if you add sharex=True we can remove the xlabel="Age" argument

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"], 
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Maximum Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate");

# Change ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
