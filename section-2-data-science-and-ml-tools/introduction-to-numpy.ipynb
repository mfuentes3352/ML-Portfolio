## Dot product example (nut butter sales)

```python
np.random.seed(0)
# Numbers of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Create weekly_sales DataFrame
import pandas as pd
weekly_sales = pd.DataFrame(sales_amounts,
                           index=["Mon","Tues","Wed","Thurs","Fri"],
                           columns=["Almond butter", "Peanut butter", "Cashew butter"])
weekly_sales
```

```python
# Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
# Create butter_prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index=["Price"], 
                            columns=["Almond butter", "Peanut butter", "Cashew butter"]) 
# Pandas is expecting 3 columns (Almond butter, Peanut butter, Cashew butter) and 1 row (Price). but prices shape is not (3,1)
# so we have to reshape it so that pandas can use this Price array to create de DataFrame 
butter_prices
```

```python
total_sales = np.dot(prices.reshape(1,3), sales_amounts.T)
#total_sales = prices.dot(sales_amount.T)
total_sales 
```

```python
weekly_sales["Total"] = total_sales.T
weekly_sales
```

```python
weekly_sales.drop(columns=["Total"], inplace=True)
```

```python
weekly_sales
```

```python
# Create daily_sales
butter_prices.shape, weekly_sales.shape
```

```python
weekly_sales.T.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
